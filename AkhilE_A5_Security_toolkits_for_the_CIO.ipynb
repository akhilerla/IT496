```python
import seaborn as sns
import pandas as  pd
import numpy as np
import matplotlib as plt
riskDF = pd.DataFrame(columns=['risk','likelihood','impact','score'])
riskDF['risk']= ['Breach of customer social security numbers',
                 'Breach of customer transactions',
                 'Breach of encrypted customer data',
                 'Breach of customer plain text password'
                 ,'Breach of customer credit card information']
riskDF['likelihood'] = [0.2,0.4,0.6,0.05,0.4]
riskDF['impact'] = [1.6,2,1.2,0.5,2.4]
riskDF['score']=riskDF['impact'] * riskDF['likelihood']
scoresPivot =  riskDF.pivot('impact','likelihood','score')
print(scoresPivot)
labelsPivot =  riskDF.pivot('impact','likelihood','risk')
print(labelsPivot)
sns.set(rc={'figure.figsize':(24,12)})
p1 = sns.heatmap(scoresPivot,cmap="cool",annot=labelsPivot,fmt="")
```
